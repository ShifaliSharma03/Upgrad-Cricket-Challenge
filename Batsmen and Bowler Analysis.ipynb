```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
ds_India=pd.read_csv('Against India.csv')[:50]
ds_Aus=pd.read_csv('Against Australia.csv')[:50]
ds_complete=ds_India.append(ds_Aus)
ds_complete['Team']=ds_complete['Player'].apply(lambda x: x[x.find("(")+1:x.find(")")])
```

```python
ds_complete=ds_complete[(ds_complete['Team']=='INDIA') | (ds_complete['Team']=='AUS') ]
```

```python
ds_complete=ds_complete.sort_values(by=['Ave'], ascending=False)
```

```python
ds_complete['Player'].unique()
```

```python
squad=['RG Sharma (INDIA)', 'V Kohli (INDIA)','MS Dhoni (INDIA)','S Dhawan (INDIA)','SE Marsh (AUS)','AJ Finch (AUS)', 
       'GJ Maxwell', 'B Kumar', 'RA Jadeja','PJ Cummins','Mohammed Shami', 'YS Chahal','JJ Bumrah']
ds_complete['In Squad']=ds_complete['Player'].apply(lambda x:'Yes' if x in squad else 'No')
ds_complete[ds_complete['In Squad']=='Yes']
```

# Based on Average against Other Team Rohit Sharma will be the Top Run Scorer from India and Shaun Marsh from Australia
# Series Top Run Scorer will be Rohit Sharma
# Most 4s V Kohli 
# Most 6s Rohit Sharma
